## Create mapper.py

```python
%%writefile mapper.py
#!/usr/bin/env python

import sys

# Input comes from STDIN (standard input)
for line in sys.stdin:
    # Extract date and temperature
    date, temperature = line.strip().split(',')

    # Determine if it's sunny or cool
    if int(temperature) >= 25:
        weather = 'sunny'
    else:
        weather = 'cool'

    # Output date and weather condition in tabular format
    print(f"{date}\t{weather}")
```

## Create reducer.py

```python
%%writefile reducer.py
#!/usr/bin/env python

import sys

# Initialize variables
current_date = None
current_weather = None

# Print column names
print("{:<15} {:<15}".format("Date", "Weather"))
print()

# Input comes from STDIN
for line in sys.stdin:
    # Extract date and weather from input
    date, weather = line.strip().split('\t', 1)

    # If the current date is different from the date we just read,
    # and if it's not the first date (i.e., current_date is not None),
    # print out the date and weather condition
    if current_date and current_date != date:
        print(f"{current_date}\t{current_weather}")

    # Update current_date and current_weather
    current_date = date
    current_weather = weather
```
